```python
# для понятия цикла for
layer = 5
P = []

for i in range(layer):
    row = [1] * (i + 1) # row = [i]... не верно...
    for j in range(i+1):
        if j != 0 and j != i:
            row[j] = P[i-1][j-1] + P[i-1][j]
            
    P.append(row)
        
for i in range(layer):
    numbers = [f"{num:3}" for num in P[i]]
    space = ' ' * (layer - i - 1) * 2  
    print(space + ' '.join(numbers))
    
    
    
# for i in range(layer): старая версия
#     space = ' ' * (layer - i)
    
#     if i % 2 == 1 and layer % 2 == 0:
#         space += ' '
#     elif i % 2 == 0 and layer % 2 == 1:
#         space += ' '
        
#     print(space + ' '.join(str(num) for num in P[i])) 
```
