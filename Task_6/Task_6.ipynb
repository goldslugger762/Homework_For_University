```python
# дополнить результат task2 предыдущей лекции (фибоначчи)

# самописным кэшем и проверить на больших числах


# для себя сравните скорости выполнения fib_cached и fib_regular

# на больших числах (больше 60)

CACHE = {} # ключ-значение

def fib_cached(n: int) -> int:
    if n in CACHE:
        return CACHE[n]
    
    if n == 1:
        result = 0
    elif n == 2:
        result = 1
    else:
        result = fib_cached(n - 1) + fib_cached(n - 2)
    
    CACHE[n] = result
    return result

print(fib_cached(100))
# для себя сравните скорости выполнения fib_cached и fib_regular -> сравнил, до этого fib(65) не выполнился за 10 минут и комп жестко гудел, а fib_cached(70) выполнился за 0.0 секунд и всё четко
```
