```python
from rich import print
from rich.console import Console

console = Console()

print("[bold yellow]Привет!\nСейчас мы сыграем с тобой в игру!\nЗагадай [bold purple]число!\n")

number_range = input("Введите диапазон и конкретное число(например -1000  1000 0) :")
parts = number_range.split()
low = int(parts[0])
high = int(parts[1])
definite = int(parts[2])

print(number_range)
print("[bold yellow]А теперь приступим!\n[bold green4]буду угадывать от {low} до {high}")
print("[bold yellow]Отвечай: [bold red]'<'[/bold red] если твое число меньше, [bold red]'>'[/bold red] если больше, [bold red]'='[/bold red] если угадал\n")
while low <= high:
    mid = (low + high) // 2
    answer = console.input(f"[bold yellow]Твоё число <, > или = {mid}\n")

    if mid == definite:
        print("[bold green4]Ура!\nТы победил!")
        break
    if answer == "=":
        print("[bold green4]Ура!\nТы победил!")
        break
    elif answer == "<":
        high = mid - 1
        print(f"[blue]Сужаю диапазон: [{low}, {high}]")
    elif answer == ">":
        low = mid + 1
        print(f"[blue]Сужаю диапазон: [{low}, {high}]")
    elif answer.lower() == "exit":
        print("[bold blue]Игра завершена!")
        break
```
