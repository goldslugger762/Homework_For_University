```python
def fib(n: int) -> int:
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)

print(fib(15))    
print(fib(30))
print(fib(45))

n = input("Введите номер числа фибоначчи: ")
n = int(n)
print(f'{n} ое число фиббоначи равно {fib(n)}')
```
